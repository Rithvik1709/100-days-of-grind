```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso
```

```python

# Load dataset
data = pd.read_csv("dataset.csv")
X = data["X"].values.reshape(-1, 1)
y = data["Y"].values

# Polynomial transformation
degree = 3
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

# Ridge Regression (L2 Regularization)
ridge = Ridge(alpha=1.0)  # Alpha controls regularization strength
ridge.fit(X_poly, y)
y_ridge_pred = ridge.predict(X_poly)

# Lasso Regression (L1 Regularization)
lasso = Lasso(alpha=0.1)  # Alpha controls sparsity
lasso.fit(X_poly, y)
y_lasso_pred = lasso.predict(X_poly)

# Visualization
plt.scatter(X, y, color="blue", label="Original Data")
plt.plot(X, y_ridge_pred, color="green", label="Ridge Regression")
plt.plot(X, y_lasso_pred, color="orange", label="Lasso Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Ridge & Lasso Regression")
plt.show()
```
