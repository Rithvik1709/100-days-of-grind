```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, MeanShift
from sklearn.preprocessing import StandardScaler

```

```python
# Load the dataset
df = pd.read_csv("blobs_clustering_dataset.csv")
X = df[["Feature_1", "Feature_2"]]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

# Apply Mean Shift
meanshift = MeanShift()
ms_labels = meanshift.fit_predict(X_scaled)

# Plotting results
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# DBSCAN Plot
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=db_labels, cmap='Set1')
axes[0].set_title("DBSCAN Clustering")

# Mean Shift Plot
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=ms_labels, cmap='Set2')
axes[1].set_title("Mean Shift Clustering")

plt.tight_layout()
plt.show()
```
