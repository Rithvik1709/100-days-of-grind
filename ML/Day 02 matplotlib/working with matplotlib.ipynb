# Types of Plots in Matplotlib

# Basic Plot in Matplotlib

A **Basic Plot** is the simplest form of data visualization that represents the relationship between two variables using a continuous line or markers. It helps in understanding trends, patterns, and variations in data over a given range.


````python
x = np.linspace(0,15,150)  
plt.plot(np.sin(x),label="Sine Wave",color="black")
plt.plot(np.cos(x),label="Cosine Wave",color="blue")
plt.legend(loc="upper center")
plt.show()
````

# Scatter Plot in Matplotlib

A **Scatter Plot** is used to display individual data points as markers, showing relationships and distributions between two continuous variables. It is useful for identifying correlations, clusters, and outliers in datasets.


````python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='blue', marker='x')
plt.title("Scatter Plot")
plt.show()
````

# Bar Chart in Matplotlib

A **Bar Chart** represents categorical data using rectangular bars, where the length or height of each bar corresponds to the value of the category. It is useful for comparing different groups or categories visually.


````python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 8]

plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])
plt.title("Bar Chart")
plt.show()
````

````python
plt.barh(categories, values, color=['red', 'blue', 'green', 'purple'])
plt.title("Horizontal Bar Chart")
plt.show()
````

# Histogram in Matplotlib

A **Histogram** represents the distribution of a dataset by dividing it into bins and counting the number of observations in each bin. It is useful for visualizing frequency distributions, detecting skewness, and understanding data spread.


````python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.show()
````

# Pie Chart in Matplotlib

A **Pie Chart** is used to represent categorical data as slices of a circle, where each slice corresponds to a proportion of the whole. It is useful for visualizing percentage distributions and comparing relative sizes of different categories.


````python
marks = np.array([92,91,90,85,81])
subjects = ["Computer","Maths","Electronics","English","Management"]
plt.pie(marks,labels=subjects,shadow=True,explode=[0,0,0.2,0,0.15],autopct="%0.3f%%")
plt.show()
````

# Subplots in Matplotlib

**Subplots** allow multiple plots to be displayed within a single figure, enabling comparative analysis of different datasets. They help organize visualizations in a grid layout, making it easier to analyze multiple relationships simultaneously.


````python
fig, ax = plt.subplots(2, 2)

x = np.linspace(0, 10, 100)
y1, y2, y3, y4 = np.sin(x), np.cos(x), np.tan(x), np.exp(-x)

ax[0, 0].plot(x, y1)
ax[0, 0].set_title("Sine Wave")

ax[0, 1].plot(x, y2)
ax[0, 1].set_title("Cosine Wave")

ax[1, 0].plot(x, y3)
ax[1, 0].set_title("Tangent Wave")

ax[1, 1].plot(x, y4)
ax[1, 1].set_title("Exponential Decay")

plt.tight_layout()
plt.show()
````

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linestyle='dashed', linewidth=2, color='red', marker='o', markersize=5, label='Sine Wave')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
````

# 3D Plot in Matplotlib

A **3D Plot** is used to visualize data in three dimensions, providing depth to the graphical representation. It is useful for analyzing relationships between three variables and includes surface plots, wireframes, and 3D scatter plots.


````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')

plt.title("3D Surface Plot")
plt.show()
````

# Saving Plots in Matplotlib

Matplotlib allows saving plots in various formats such as PNG, JPG, PDF, and SVG. Saving plots ensures reproducibility and sharing of visualizations. Users can specify the filename, resolution (DPI), and format while saving.


````python
plt.plot(x, y)
plt.title("Saving Figure Example")
plt.savefig("plot.png", dpi=300)
plt.show()
````

# Box Plot in Matplotlib

A **Box Plot** (or **Box-and-Whisker Plot**) is used to display the distribution of a dataset through five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It helps in identifying outliers, skewness, and variability in the data.


````python
data = [np.random.rand(50), np.random.rand(50) * 2, np.random.rand(50) * 3]
plt.boxplot(data)
plt.title("Boxplot Example")
plt.show()
````
